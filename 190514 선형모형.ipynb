```python
import pandas as pd
mise=pd.read_csv('C:\\Users\\user\\Desktop\\mises.csv',encoding='cp949')

meteo=pd.read_csv('C:\\Users\\user\\Desktop\\meteos.csv',encoding='cp949')

fin=pd.merge(mise,meteo,on=['지역','날짜'])

data=fin['날짜'].astype(str)
fin['일시'] = data.str.slice(start=0, stop=8).astype(str)

fin['일시']=pd.to_datetime(fin['일시'])
fin.rename(columns={'지역':'시도'},inplace=True)

fin.일시.dt.weekday

fin.fillna(0,inplace=True)

fin['요일'] = fin.일시.dt.weekday
```

```python
fin.set_index('일시', inplace=True)
```

```python
del fin['Unnamed: 0']
del fin['Unnamed: 0.1']
```

```python
fin_1=fin.shift(1)
```

```python
fin.reset_index('일시',inplace=True)
fin_1.reset_index('일시',inplace=True)
```
